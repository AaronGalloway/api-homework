```python
# Import libraries and dependencies
import os
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation
```

```python
# Load .env enviroment variables
load_dotenv(dotenv_path='example.env')
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv('ALPACA_KEY')
alpaca_secret_key = os.getenv('ALPACA_SECRET_KEY')
```

```python
# Create the Alpaca API object
alpaca = tradeapi.REST(alpaca_api_key,alpaca_secret_key, api_version="v2")
```

### Get 5 years of historical closing prices fora  traditional 40/60 portfolio. 40% AGG and 60% SPY

```python
# Set timeframe to '1D'
timeframe = "1D"

# Set start and end datetimes between now and 5 years ago. October 1, 2015 to October 1, 2020
start_date = pd.Timestamp("2015-10-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2020-10-01", tz="America/New_York").isoformat()

# Set the ticker information
tickers = ["SPY","AGG"]

# Get historical price data
portfolio_dataframe = alpaca.get_barset(tickers, timeframe, start = start_date, end=end_date).df
portfolio_dataframe

# Display sample data
portfolio_dataframe.head()
```

```python
# Configure a Monte Carlo simulation to forecast five years cumulative returns. 500 runs for 30 years at a 40/60 split
weights = [.40,.60]
MC_Simulation = MCSimulation(portfolio_data = portfolio_dataframe, weights=weights,num_simulation=500, num_trading_days=252*30)

# Print the simulation input data
MC_Simulation.portfolio_data.head()
```

```python
# Run a Monte Carlo simulation to forecast five years cumulative returns
MC_Simulation.calc_cumulative_return()
```

```python
# Plot simulation outcomes
MC_Simulation.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
MC_Simulation.plot_distribution()
```

```python
# Fetch summary statistics from the Monte Carlo simulation results
Summary = MC_Simulation.summarize_cumulative_return()
Summary
```

```python
# Given an initial investment of $20,000 what is the expected portfolio return in dollars at the 95% lower and uppoer confidence ratio
initial_investment = 20000
lower = round((initial_investment * Summary[8]),2)
higher = round((initial_investment * Summary[9]),2)
print(f'There is a 95% chance that an initial investment of ${initial_investment} in the portfolio over the next 5 years will end within the range of ${lower} and ${higher}')
```

```python
# Given a 50% increase in the initial investment what would be the return at the 95% lower and upper interval
bigger_initial_investment = 30000
lower = round((bigger_initial_investment * Summary[8]),2)
higher = round((bigger_initial_investment * Summary[9]),2)
print(f'There is a 95% chance that an initial investment of ${bigger_initial_investment} in the portfolio over the next 5 years will end within the range of ${lower} and ${higher}')
```
